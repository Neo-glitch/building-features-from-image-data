**Aim is to learn how to preserve image keypoints when we perform image augmentaion on an image**

```python
# 7

import matplotlib.pyplot as plt
import albumentations as albu
from skimage import io
import numpy as np
import cv2
import os
```

```python
peeing_dog = io.imread("./datasets/images/dog_face.jpg")
```

```python
# array to hold cords of keypoints in image manually, normally we have algo for this
points = [(120, 1750), (470, 1560), (800, 1460), (1250, 1200), (1540, 1150), (1840, 1200),
         (2520, 1180), (2540, 1140), (3000, 1520), (3560, 1700), (820, 1750), (1300, 1700),
         (2350, 1720), (2800, 1750), (270, 2300), (1200, 2300), (1750, 2000), (2300, 2300), (3400, 2300)]
```

```python
# helper fn to visualize the keypoint on input image
def visualize_points(peeing_dog, points, diameter = 24):
    im = peeing_dog.copy()
    
    for(x, y) in points: # iters through manual keypoint and plot circle at pos
        cv2.circle(im, (int(x), int(y)), diameter, (0, 0, 255), -1)
        
    plt.figure(figsize=(6,6))
    plt.imshow(im)
```

```python
# fn does image transforms whilst keeping the image keypoints
def transformer(transformations):
    
    # returns the image as well as the keypoints
    return albu.Compose(transformations, p  = 1,
                        # xy is format of keypoints i.e x, y
                       keypoint_params={"format": "xy"})(image=peeing_dog, keypoints = points)
```

```python
visualize_points(peeing_dog, points)
```

```python
# Vertical flip transform
transformed_img = transformer([albu.VerticalFlip(p = 1)])

key_points = transformed_img["keypoints"]   # get the keypoints preserved in transformed image
im = transformed_img["image"]   # gets transformed_image

visualize_points(im, key_points)
```

```python
# crop image
transformed_img = transformer([albu.RandomCrop(width = 1024,
                                              height = 1024, p = 1)])

key_points = transformed_img["keypoints"]   # get the keypoints preserved in transformed image
im = transformed_img["image"]   # gets transformed_image

visualize_points(im, key_points)
```

```python
# rotate image
transformed_image = transformer([albu.RandomRotate90(p=1)])

key_points = transformed_image["keypoints"]   # get the keypoints preserved in transformed image
im = transformed_image["image"]   # gets transformed_image

visualize_points(im, key_points)
```

```python
# rotate image at any specified degree
transformed_image = transformer([albu.Rotate(p=1)])

key_points = transformed_image["keypoints"]   # get the keypoints preserved in transformed image
im = transformed_image["image"]   # gets transformed_image

visualize_points(im, key_points)
```

```python
# crop image
transformed_image = transformer([albu.CenterCrop(height = 512, width = 512, p=1)])

key_points = transformed_image["keypoints"]   # get the keypoints preserved in transformed image
im = transformed_image["image"]   # gets transformed_image

visualize_points(im, key_points)
```

```python
# crop image
transformed_image = transformer([albu.RandomSizedCrop(
    min_max_height = (256, 1024),
    height = 512, width = 512, p=1)])

key_points = transformed_image["keypoints"]   # get the keypoints preserved in transformed image
im = transformed_image["image"]   # gets transformed_image

visualize_points(im, key_points)
```

```python
# Series of transformation, all at once
transformed_image = transformer([albu.CenterCrop(height = 512, width = 512, p=1),
                                albu.ShiftScaleRotate(p=1),
                                albu.OneOf([albu.HueSaturationValue(p = 0.5),
                                           albu.RGBShift(p = 0.7)], p =1)])

key_points = transformed_image["keypoints"]   # get the keypoints preserved in transformed image
im = transformed_image["image"]   # gets transformed_image

visualize_points(im, key_points)
```
