```python
# 3

import skimage
from skimage import data
from skimage import io

import matplotlib.pyplot as plt
import numpy as np
```

```python
cat = io.imread("./datasets/images/cat.jpg")

plt.figure(figsize = (6, 6))
plt.imshow(cat)
```

```python
# resized cat image
from skimage.transform import resize

# anti-aliasing helps min distortion when scaling from high res to lower res
cat_resized = resize(cat, (480, 480), anti_aliasing=True)  
cat_resized.shape
```

```python
plt.figure(figsize = (6, 6))
plt.imshow(cat_resized)

# image looks strange, since we didn't use aspect ratio of image when scaling down
```

```python
# aspect ratio = width / height

aspect_ratio_original = cat.shape[1] / float(cat.shape[0])
aspect_ratio_resized = cat_resized.shape[1] / float(cat_resized.shape[0])

print("Original aspect ratio: ", aspect_ratio_original)
print("resized aspect ratio: ", aspect_ratio_resized)
```

```python
# scale image and also preserves aspect ratio
from skimage.transform import rescale

# scale image with height same and height to be scaled wrt original aspect ratio
cat_rescaled = rescale(cat_resized, (1.0, aspect_ratio_original), multichannel=True)
cat_rescaled.shape
```

```python
plt.figure(figsize = (6, 6))
plt.title("Rescaled wrt aspect ratio")
plt.imshow(cat_rescaled)
```

```python
# helper fn that crops image at center
def crop_center(img, cropx, cropy):
    height, width, channels = img.shape
    
    start_width = width // 2 - (cropx // 2)  # starts crop at center of image
    start_height = height // 2 - (cropy // 2)
    
    return img[start_height:start_height + cropy, start_width:start_width + cropx]
```

```python
cat_cropped = crop_center(cat_resized, 244, 244)

plt.figure(figsize = (10, 10))
plt.imshow(cat_cropped);
```

```python
from skimage import color

cat_grey = color.rgb2gray(cat)
cat_grey.shape
```

```python
from skimage.transform import resize

cat_resized = resize(cat_grey, (344, 516), anti_aliasing=True)
cat_resized.shape
```

```python
# size of blockView that's used for subsampling part of the image or size of filter(kinda like cnn pooling)
block_shape = (4, 4)
```

```python
from skimage.util import view_as_blocks

cat_blocks = view_as_blocks(cat_resized, block_shape)
```

```python
cat_blocks.shape
```

```python
# flattened the block to be in shape 86, 129, 16
flattened_blocks = cat_blocks.reshape(cat_blocks.shape[0], cat_blocks.shape[1], -1)

print("Shape of the blocks image: ", cat_blocks.shape)
print("Shape of the flattend image: ", flattened_blocks.shape)
```

```python
# op is similar to avg pooling when creating NN with tf
mean_blocks = np.mean(flattened_blocks, axis = 2)

plt.figure(figsize = (10, 10))
plt.imshow(mean_blocks, interpolation = "nearest", cmap = "gray")
```

```python
# op is similar to max pooling when creating NN with tf
max_blocks = np.max(flattened_blocks, axis = 2)

plt.figure(figsize = (10, 10))
plt.imshow(max_blocks, interpolation = "nearest", cmap = "gray")
```
