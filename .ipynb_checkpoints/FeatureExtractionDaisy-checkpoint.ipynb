```python
# 9th

import skimage
from skimage import color
from skimage.feature import daisy
from skimage import data
from skimage.transform import resize
import matplotlib.pyplot as plt
import cv2
```

```python
horse_gray = cv2.imread("./datasets/images/horse.jpg", 0)

horse = resize(horse_gray, output_shape = (300, 450), mode = "reflect", anti_aliasing=False)

plt.figure(figsize = (6,6))
plt.imshow(horse, cmap ="gray")
```

```python
# gets descriptor array
desc_array, desc_image = daisy(horse, step = 100, # steps is dist in pixel btw desc sampling points of algo
                               radius = 30, visualize = True)
```

```python
plt.figure(figsize = (9, 9))
plt.title("Daisy descriptors", fontsize = 10)
plt.imshow(desc_image)
```
