**Aim of notebook is just manipulation of colorchannel of an image**

```python
#2

import skimage
from skimage import data
from skimage import io

import numpy as np
import matplotlib.pyplot as plt
```

```python
astronaut = data.astronaut()

print(astronaut.shape)
plt.figure(figsize = (6, 6))
plt.imshow(astronaut)
```

```python
blue_astronaut = astronaut.copy()

blueish = astronaut[:, :, 0] > 160  # gets loc in image where r color channel value > 160
blue_astronaut[blueish] = [0, 0, 255]


plt.figure(figsize =(6, 6))
plt.imshow(blue_astronaut)
```

```python
# manip using util lib
from skimage import util
```

```python
inverted_astronaut = util.invert(astronaut)  # inverts intensity range of image, max intensity is min and viceversa

plt.figure(figsize = (6, 6))
plt.imshow(inverted_astronaut)
```

```python
# switching from RGB to BGR color channel
astronaut = data.astronaut()

BGR_astronaut = astronaut[:, :, ::-1]
plt.figure(figsize = (6, 6))
plt.imshow(BGR_astronaut)
```

```python
# converts colored image to grayscale
from skimage.color import  rgb2gray
```

```python
balloon = io.imread("./datasets/images/balloons.jpg")

plt.figure(figsize = (6, 6))
plt.imshow(ballon)
```

```python
gray_balloon = rgb2gray(balloon)
plt.figure(figsize = (6, 6))
plt.imshow(gray_balloon, cmap = "gray")
```

```python
# converts rgb to cielab color space
from skimage.color import rgb2lab

lab_image = rgb2lab(balloon)

plt.figure(figsize = (6, 6))
plt.imshow(lab_image.astype(np.uint8))
```
