```python
from skimage import io
import numpy as np
import cv2
import matplotlib.pyplot as plt

import albumentations as albu
```

```python
image = io.imread("./datasets/images/taj_mahal_india.jpg")

plt.figure(figsize = (6, 6))
plt.imshow(image)
```

```python
# helper fn to augment and show the augmented image
def augment_and_show(aug, image):
    image = aug(image = image)["image"]  # use augmentation function on image and ret the image
    plt.figure(figsize = (6, 6))
    plt.imshow(image)
    
    
img_rain = albu.RandomRain(p=1, brightness_coefficient=0.9, drop_width = 3, blur_value = 1)
augment_and_show(img_rain, image)
```

```python
# augment with snow
img_snowfall = albu.RandomSnow(p = 1, brightness_coeff = 1.9, snow_point_lower = 0.3, snow_point_upper = 0.7)

augment_and_show(img_snowfall, image)
```
