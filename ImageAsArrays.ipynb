```python
# reading real image from pc
chessboard = io.imread("./datasets/images/chessboard.jpg")

print("Type: ", type(chessboard))
print("dtype: ", chessboard.dtype)
print("shape: ", chessboard.shape)  # has col for color channel i.e 3
print("size: ", chessboard.size) 
    
plt.figure(figsize = (7, 7))
plt.imshow(chessboard, cmap = "gray")
```

**changing image mat data type**

```python
from skimage import img_as_ubyte, img_as_float  # change image matrix type maybe from int to float and more

image = data.coffee()
plt.figure(figsize = (6, 6))
plt.imshow(image)
```

```python
image_ubyte = img_as_ubyte(image)
image_float = img_as_float(image)
```

```python
# gets min and max values(pixels) of the converting image
print("type, min, max: ", image_ubyte.dtype, image_ubyte.min(), image_ubyte.max())
print("type, min, max: ", image_float.dtype, image_float.min(), image_float.max())
```

```python
# gets individual color channels of a colored image
image = io.imread("./datasets/images/machu_pichho_peru.jpg")

r = image[:, :, 0]  # image with just red color channel
g = image[:, :, 1]
b = image[:, :, 2]
```

```python
fig, ax = plt.subplots(1, 4, figsize = (16, 5))

(ax_r, ax_g, ax_b, ax_color) = ax

ax_r.imshow(r, cmap = "gray")
ax_r.set_title("ch red")

ax_g.imshow(g, cmap = "gray")
ax_g.set_title("ch green")

ax_b.imshow(b, cmap = "gray")
ax_b.set_title("ch blue")

# stack the color channels back and show normal colored image
ax_color.imshow(np.stack([r, g, b], axis = 2))
ax_color.set_title("all channels")
```
