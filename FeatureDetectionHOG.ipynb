```python
# 10th

import matplotlib.pyplot as plt
import skimage
from skimage.feature import hog
from skimage import data, io, color, exposure
```

```python
man_img = skimage.io.imread("./datasets/images/crossing.jpg")

plt.figure(figsize= (6, 6))
plt.imshow(man_img)
```

```python
# crop image to smaller portion
from skimage.util import crop

cropped_img = crop(man_img, ((1500, 600), (700, 700), (0, 0)), copy = False)
plt.figure(figsize= (6, 6))
plt.imshow(cropped_img)
```

```python
# RESIZR image again to speed up preprocessing
from skimage.transform import resize

resized_img = resize(cropped_img, (256, 256), mode = "reflect", anti_aliasing = True)
resized_img.shape
```

```python
# hog feature extraction
# ret feature dectectors as well viz of resulting image
fd, hog_image = hog(resized_img,
                   pixels_per_cell=(6,6),   # smaller block sizes give more detailed descriptors
                   block_norm = "L2-Hys",  # tries to make gradient calc invariate to light change
                   visualize = True,    # ret viz of resulting image
                   multichannel = True)
```

```python
fig, axes = plt.subplots(1, 2,
                        figsize = (10, 10), sharex=True, sharey=True)

ax = axes.ravel()

ax[0].imshow(resized_img, cmap = "gray")
ax[0].set_title("Original")

# shrinks or strenthens intensity range of output hog image
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

ax[1].imshow(hog_image_rescaled, cmap = "gray")
ax[1].set_title("HOG_image")

plt.tight_layout()
plt.show()
```
